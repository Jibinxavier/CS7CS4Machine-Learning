```r
sum_dataset <- read.table("data/without noise/The SUM dataset, without noise.csv", header=TRUE, sep=";", stringsAsFactors=FALSE, nrows=100000)
head(sum_dataset)
```

## Split the dataset

## Set Up Evaluaion Method

```r
# Function that returns Root Mean Squared Error
rmse <- function(error)
{
  sqrt(mean(error^2))
}

# Function that returns Mean Absolute Error
mae <- function(error)
{
  mean(abs(error))
}
```

## Set Up Fits algorihm

```r
fitsRandomForest <- function(fmla, dataset ){
    randomForest(fmla, data = dataset)
}
```
